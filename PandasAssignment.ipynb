````python
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams.update({'font.size': 20, 'figure.figsize': (10, 8)})
````

````python
df = pd.read_csv("travel-times.csv", index_col=0)
df
````

````python
print(df.shape)
````

````python
df.head(15)
````

````python
missing_info = df.isnull().sum()
print("List of columns with missing values\n (True means missing at-least one value;\n false means nothing is missing):\n",missing_info > 0)
````

````python
df.drop_duplicates(inplace=True,keep='last')
df
````

````python
missing_info = df.isnull().sum()
print("Number of missing values in each column\n (0 indicates no missing values):\n",missing_info)
````

````python
df_resolved_rows = df.copy()
df_resolved_columns = df.copy()
````

````python
df_resolved_rows.dropna(inplace=True)
df_resolved_rows
````

````python
df_resolved_columns.dropna(axis=1,inplace=True)
df_resolved_columns
````

````python
print(df.describe())
print("Correlation Matrix:\n", df.corr(numeric_only=True))
````

````python
df.iloc[11:15]
````

````python
subset = df[(df.index == '11/23/2011') | (df.index == '1/6/2012')]
subset
````

````python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20, 'figure.figsize': (10, 8)})
df.plot(kind = 'scatter', x = 'TotalTime', y = 'Distance', title = 'Distance vs TotalTime')
````
